Automobile Sales Forecasting Using Functional Data Analysis

```python
import pandas as pd
import numpy as np
import skfda
import matplotlib.pyplot as plt
from skfda.ml.regression import KNeighborsRegressor
from scipy.stats import gaussian_kde
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from skfda.misc.metrics import l2_distance, l2_norm
from skfda.exploratory.visualization import FPCAPlot
from skfda.preprocessing.dim_reduction.feature_extraction import FPCA
from math import sqrt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
#Figure 1: cumulative percentages of variance explained by different numbers of functional principal components
```

```python
data_matrix_train = pd.read_csv('data_matrix_train.csv')
del data_matrix_train['Unnamed: 0']
data_matrix_test = pd.read_csv('data_matrix_test.csv')
del data_matrix_test['Unnamed: 0']
y = pd.concat([data_matrix_train,data_matrix_test], axis=1)
y.columns = range(1,41)
grid_points = []
for i in range(1,41):
    grid_points.append(i)
fd_y = skfda.FDataGrid(
    data_matrix=y,
    grid_points=grid_points,
)
fpca = FPCA(n_components=15)
fpca.fit(fd_y)
ratio = fpca.explained_variance_ratio_
cum_ratio = [0] #累计百分比
for i in range(0,15):
    a = 0
    for j in range(0,i+1):
        a = a + ratio[j]
    cum_ratio.append(a)
x_axes = range(0,16)
y_axes = cum_ratio
plt.figure(figsize=(10,5))
plt.plot(x_axes,y_axes,marker='o')
plt.xticks(x_axes)
plt.xlabel('Number of Functional Principal Components')
plt.ylabel('Cumulative Percentages of Variance Explained')
plt.title('Functional Principal Components Analysis')
plt.plot([0, 15], [0.90, 0.90], dashes=[2, 1])
plt.show()
```

```python
#Figure 2: densities of reconstruction errors (RE) with 4 functional principal components
```

```python
q = 4
fpca_clean = FPCA(n_components=q)
fpca_clean.fit(fd_y)
fd_y_hat = fpca_clean.inverse_transform(
    fpca_clean.transform(fd_y)
)
err_fd_y = l2_distance(
    fd_y,
    fd_y_hat
) / l2_norm(fd_y)
x_density = np.linspace(0., 1.6, num=2276)
density_err_fd_y = gaussian_kde(err_fd_y)
err_thresh = np.quantile(err_fd_y, 0.95)
density_fd_y_err_outl = gaussian_kde(err_fd_y[err_fd_y >= err_thresh])
density_fd_y_err_inli = gaussian_kde(err_fd_y[err_fd_y < err_thresh])
plt.figure(figsize=(10,5))
plt.plot(
    x_density,
    density_err_fd_y(x_density),
    label='Error functional data',
    color='grey'
)
plt.vlines(
    err_thresh,
    ymax=max(density_fd_y_err_outl(x_density)),
    ymin=0.0,
    label='thresh=quantile(p=0.95)',
    linestyles='dashed',
    color='black'
)
plt.plot(
    x_density,
    density_fd_y_err_outl(x_density),
    label='Error test>= thresh (outliers)',
    color='C1'
)
plt.plot(
    x_density,
    density_fd_y_err_inli(x_density),
    label='Error test< thresh (nonoutliers)',
    color='red'
)
plt.xlabel('Relative L2-norm reconstruction errors')
plt.ylabel('Density (unnormalized)')
plt.title(f'Densities of reconstruction errors with {q} components')
plt.legend(loc='best')
plt.show()
```

```python
#Figure 3: original data and smoothed data with different basis functions
```

```python
newsales = y[err_fd_y < err_thresh]
fd_y = skfda.FDataGrid(
    data_matrix=newsales,
    grid_points=grid_points,
)
fourier_basis = skfda.representation.basis.VectorValued([
    skfda.representation.basis.Fourier(  # Fourier
        n_basis=25,
        domain_range=fd_y.domain_range,
    ),
])
fd_y_fourier = fd_y.to_basis(fourier_basis)
bspline_basis = skfda.representation.basis.VectorValued([
    skfda.representation.basis.BSpline(  # BSpline
        n_basis=50,
        domain_range=fd_y.domain_range,
    ),
])
fd_y_bspline = fd_y.to_basis(bspline_basis)
fig, axes = plt.subplots(1, 3, figsize=(15, 3)) 
fd_y[0:10].plot(axes=axes[0])
axes[0].set_title('Original')
fd_y_fourier[0:10].plot(axes=axes[1])
axes[1].set_title('25 Fourier basis functions')
fd_y_bspline[0:10].plot(axes=axes[2])
axes[2].set_title('50 b-spline basis functions')
plt.show()
```

```python
#Figure 4: original data and predicted data
```

```python
train, test = train_test_split(newsales, test_size = 0.2, random_state=10)
X_train = train.iloc[:,0:36]
y_train = train.iloc[:,36:40]
X_train = skfda.FDataGrid(
    data_matrix=X_train,
    grid_points=grid_points[0:36],
)
y_train = skfda.FDataGrid(
    data_matrix=y_train,
    grid_points=grid_points[36:40],
)
y_train_fourier = y_train.to_basis(fourier_basis)
knn = KNeighborsRegressor(n_neighbors=7, weights='distance')
knn.fit(X_train, y_train_fourier)
X_test = test.iloc[:,0:36]
X_test = skfda.FDataGrid(
    data_matrix=X_test,
    grid_points=grid_points[0:36],
)
y_pred = knn.predict(X_test)
y_test = test.iloc[:,36:40]
y_test = skfda.FDataGrid(
    data_matrix=y_test,
    grid_points=grid_points[36:40],
)
y_test_fourier = y_test.to_basis(fourier_basis)
pred = y_pred.to_grid(grid_points = grid_points[36:40])
fig1 = plt.figure(figsize=(15,5))
y_test_fourier[[0,2,3,4,5,6,7,8,9,10]].plot(fig=fig1)
fig1.axes[0].set_prop_cycle(None) 
y_pred[[0,2,3,4,5,6,7,8,9,10]].plot(fig=fig1, linestyle='--')
plt.xlim(37,40)
plt.ylim(0,200)
plt.title('Original & Predicted')
plt.ylabel('Number of sales')
plt.xlabel('January 2021 - April 2021')
plt.xticks([])
plt.show()
```

```python
cnt_submarket = pd.read_csv('cnt_submarket.csv') #细分市场数据集
data_matrix_submarket = pd.read_csv('data_matrix_submarket.csv')
del data_matrix_submarket['Unnamed: 0']
data_matrix_submarket['细分市场'] = cnt_submarket['细分市场']
PCA2 = data_matrix_submarket.groupby(by=['细分市场']).sum()
PCA2_dict = {idx: PCA2.loc[idx] for idx in PCA2.index}
def plot_submarket(submarket,PCA2,PCA2_dict,ax):
    for idx in PCA2.index:
        x = range(1,41)
        y = PCA2_dict[idx]
        if idx == submarket:
            ax.plot(x, y, color="#0b53c1", lw=2.4, zorder=10)
            ax.scatter(x, y, fc="w", ec="#0b53c1", s=60, lw=2.4, zorder=12)      
            # If not, use a gray and thinner line
        else:
            ax.plot(x, y, color="#BFBFBF", lw=1.5)
        ax.set_title(submarket, fontsize=15, fontweight=500);
typesubmarket = PCA2.index[[0,1,7,9,12,30]]
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(15, 8))
for idx, (ax, submarket) in enumerate(zip(axes.ravel(), typesubmarket)):
    plot_submarket(submarket, PCA2, PCA2_dict, ax)
plt.show()
```

```python
cnt_citylevel = pd.read_csv('cnt_citylevel.csv') #城市级别数据集
data_matrix_citylevel = pd.read_csv('data_matrix_citylevel.csv')
del data_matrix_citylevel['Unnamed: 0']
data_matrix_citylevel['城市级别'] = cnt_citylevel['城市级别']
data_matrix_citylevel.loc[data_matrix_citylevel['城市级别'] == '1级市场','城市级别']='Tier 1'
data_matrix_citylevel.loc[data_matrix_citylevel['城市级别'] == '2级市场','城市级别']='Tier 2'
data_matrix_citylevel.loc[data_matrix_citylevel['城市级别'] == '3级市场','城市级别']='Tier 3'
data_matrix_citylevel.loc[data_matrix_citylevel['城市级别'] == '4级市场','城市级别']='Tier 4'
data_matrix_citylevel.loc[data_matrix_citylevel['城市级别'] == '5级市场','城市级别']='Tier 5'
fd_y = skfda.FDataGrid(
    data_matrix=data_matrix_citylevel.iloc[:,0:40],
    grid_points=grid_points,
)
fourier_basis = skfda.representation.basis.VectorValued([
    skfda.representation.basis.Fourier(  # Fourier
        n_basis=250,
        domain_range=fd_y.domain_range,
    ),
])
fd_y_fourier = fd_y.to_basis(fourier_basis)
mean = data_matrix_citylevel.groupby(by=['城市级别']).mean()
fd_mean = skfda.FDataGrid(
    data_matrix=mean,
    grid_points=grid_points,
)
fd_mean_fourier = fd_mean.to_basis(fourier_basis)
fig, axes = plt.subplots(2, 2, figsize=(15, 8))
d1 = fd_y.derivative() 
fig1 = d1.plot(
    group=data_matrix_citylevel['城市级别'],
    linewidth=1,
    alpha=0.7,
    legend=True,
    axes=axes[0][0],
)
axes[0][0].set_title('1st derivative')
d2 = fd_y.derivative(order=2)
fig2 = d2.plot(
    group=data_matrix_citylevel['城市级别'],
    linewidth=1,
    alpha=0.7,
    legend=True,
    axes=axes[0][1],
)
axes[0][1].set_title('2nd derivative')
d3 = fd_mean.derivative() 
fig1 = d3.plot(
    group=mean.index,
    linewidth=1,
    alpha=0.7,
    legend=True,
    axes=axes[1][0],
)
axes[1][0].set_title('Mean - 1st derivative')
d4 = fd_mean.derivative(order=2) 
fig1 = d4.plot(
    group=mean.index,
    linewidth=1,
    alpha=0.7,
    legend=True,
    axes=axes[1][1],
)
axes[1][1].set_title('Mean - 2nd derivative')
plt.show()
```
